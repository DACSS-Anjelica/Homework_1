## **I. Preparation**

**I1. Load libraries and data**

```python
# Import libraries used for spatial data handling, analysis, and visualization
import geopandas as gpd           # handling geospatial vector data
import numpy as np                # numerical operations
import pandas as pd               # data manipulation
from libpysal.graph import Graph  # spatial weights / neighborhood graphs
import esda                       # exploratory spatial data analysis (LISA)
import matplotlib.pyplot as plt   # plotting

# Set a random seed to ensure reproducible results
from numpy.random import seed
seed(42)
```

**I2. Read the map file**

```python
# Save peru map file from github url to this variable
LinkPeru="https://raw.githubusercontent.com/DACSS-Anjelica/Homework_1/main/PeruMaps.gpkg"

# read the same layer used in the tutorial
peru = gpd.read_file(LinkPeru, layer='good_geom')

# Confirms it loaded correctly
peru.head()
```

```python
# Functions created to reduce repetition --------------------------------------

# Define a helper function that converts a spatial graph adjacency structure
# from long (sparse) format into a wide matrix format for inspection.
toWideMatrix = lambda g:pd.DataFrame(g.adjacency).unstack().fillna(0)

# Define a dictionary of common arguments used for interactive map visualizations,
# including the base GeoDataFrame, node styling, edge styling, and initial zoom level.
general_arguments = dict(gdf = peru,
                         node_kws = dict(color='red'),
                         edge_kws = dict(alpha=0.4, color='blue'),
                         zoom_start = 6)
```

**I3. Review the map data**

According to this output:

* the geom shape is multipolygonal
* there are 1,873 rows and 10 columns.

```python
# Check the coordinate reference system (CRS) and confirm it is EPSG:4326
peru.crs.to_epsg(),\

# Check whether the CRS is projected (False indicates a geographic CRS)
peru.crs.is_projected, \

# Identify the geometry type(s) present in the dataset (e.g., MultiPolygon)
set(peru.geom_type),\

# Display the data types of each column in the GeoDataFrame
peru.dtypes, \

# Display the shape of the dataset (number of rows and columns)
peru.shape
```

**I4. Plot Peru's map**

```python
# Plot a basic map of Peru to visually inspect the spatial extent
# and confirm that the geometries loaded correctly.
peru.plot()
```

**I5. Focus on the Lima departamiento**

```python
# Filter the dataset to include only districts belonging to the
# department of Lima for the analysis.
lima = peru[peru['DEPARTAMENTO'] == 'Lima']

# Display the first five rows to verify that the filtering was applied correctly.
lima.head()
```

## **II. Neighborhood**

### **II1. Contiguity - Queen**

**II1A. Adjacency Matrix**

```python
# Creates a spatial neighbor graph using Queen contiguity.
# Each district is connected to all districts it touches (edge or vertex).
queen_lima = Graph.build_contiguity(lima, rook = False)
```

```python
# Inspect the adjacency list showing which districts are neighbors
# under Queen contiguity (for diagnostic purposes only).
queen_lima.adjacency
```

**II1B. Adjacency Plot**

```python
# Display an interactive map to visually explore Queen contiguity relationships
# (used for exploratory purposes only, not required for analysis).
queen_lima.explore(**general_arguments)
```

### **II2. Proximity - Nearest Neighbors (k=8)**

**II2A. Adjacency Matrix**

```python
# Build a K-nearest neighbors (KNN) spatial graph using representative points
# so that each district is connected to its 8 closest neighboring districts by distance.
knn8_lima = Graph.build_knn(lima.representative_point(), # GDF
                                 k=8) # desired k

# Display the adjacency list for the KNN graph, showing which districts
# are considered neighbors under the KNN (k=8) definition.
knn8_lima.adjacency
```

**II2B. Adjacency Plot**

```python
# Display an interactive map to visually explore the KNN (k = 8) neighborhood
# structure and inspect how districts are connected under this definition.
knn8_lima.explore(**general_arguments)
```

**Spatial Lag of Queen Rule**

```python
# Display basic summary statistics for the education variable
# to inspect its distribution and check for missing values.
lima.Educ_sec_comp2019_pct.describe()
```

```python
# Plot a choropleth map of the education variable using quantile classification
# to visually explore its spatial distribution across districts.
lima.plot(
    "Educ_sec_comp2019_pct",
    scheme = "quantiles",
    cmap = "Reds_r",
    legend=True, figsize = (12, 10))
```

```python
# Builds a Queen contiguity spatial neighbor graph where districts are neighbors
# if they share a boundary or a corner.
lima_queen = Graph.build_contiguity(lima, rook=False)
```

```python
# Normalize the Queen spatial weights using row-standardization.
lima_queen = lima_queen.transform("r")
```

```python
# Compute the spatial lag of the education variable using Queen weights,
# representing the average value in neighboring districts.
ylag = lima_queen.lag(lima["Educ_sec_comp2019_pct"])
```

```python
# Add the spatial lag of the education variable as a new column
# representing the average value in neighboring districts.
lima = lima.assign(Educ_sec_comp2019_pct_lagged = ylag)

# Display the first rows to verify the new column was added.
lima.head()
```

```python
# Create a Moran scatterplot showing the relationship between each districts
# education value and the average value of its neighboring districts.
lima.plot.scatter("Educ_sec_comp2019_pct","Educ_sec_comp2019_pct_lagged")
```

**Normalization of Nearest Neighbors Rule**

```python
# Convert the row-standardized KNN spatial weights into a full matrix
# for inspection and visualization purposes.
toWideMatrix(knn8_lima.transform("r"))
```

**Spatial Lag of Nearest Neighbors Rule**

```python
# Build a K-nearest neighbors spatial graph (k = 8) using representative points
knn8_lima = Graph.build_knn(lima.representative_point(), k=8)

# Normalize the KNN spatial weights using row-standardization so that
# each districts neighbor weights sum to 1.
knn8_lima = knn8_lima.transform("r")

# Compute the spatial lag of the education variable using the KNN (k=8) weights,
# representing the average value among the nearest neighboring districts.
ylag2 = knn8_lima.lag(lima["Educ_sec_comp2019_pct"])
```

```python
# Add the spatially lagged version of the education variable (based on KNN weights)
# as a new column in the GeoDataFrame.
lima = lima.assign(Educ_sec_comp2019_pct_lagged2 = ylag2)

# Display the first few rows to verify that the new lagged variable
# was added correctly to the dataset.
lima.head()
```

```python
# Create a Moran scatterplot showing the relationship between each districts
# education value and the spatially lagged education value based on KNN (k = 8) weights.
lima.plot.scatter("Educ_sec_comp2019_pct","Educ_sec_comp2019_pct_lagged2")
```

### **Spatial Autocorrelation**

**Moran's LISA with `Lima_Queen`**

```python
# Run LISA (Local Morans I) for the education variable using Queen weights.
lisa = esda.Moran_Local(lima['Educ_sec_comp2019_pct'], lima_queen)
```

```python
# Plot the LISA cluster map using Queen weights, highlighting statistically
# significant clusters and outliers at the 0.05 significance level.
lisa.plot(lima, crit_value = 0.05, figsize = (10, 13), legend = True)
```

```python
# Get the LISA cluster (quadrant) labels for each district using Queen weights
# at the 0.05 significance level.
lima['HS_lisa_queen'] = lisa.get_cluster_labels(crit_value = 0.05)

# Display the count of districts in each LISA category
# to summarize the distribution of cluster types.
lima['HS_lisa_queen'].value_counts()
```

```python
# Define the summary statistics to compute for the education variable
TheStats = ["mean", "min", "max","var"]

# Calculate summary statistics of the education variable grouped by
# Queen-based LISA cluster category.
grouped_stats = lima.groupby('HS_lisa_queen')['Educ_sec_comp2019_pct'].agg(TheStats)

# Display the summary statistics for each LISA cluster group.
grouped_stats
```

```python
# Display the unique LISA cluster categories assigned under
# the Queen contiguity neighborhood definition.
lima['HS_lisa_queen'].unique()
```

```python
# Define the original LISA cluster labels returned by the analysis
oldLabels = ['Insignificant', 'Low-Low', 'High-High', 'High-Low', 'Low-High']

# Define new, custom labels to rename the LISA categories for clarity and ordering
# (with 'no_pattern' placed in the middle).
newLabels = ['3 no_pattern', '4 coldSpot','2 hotSpot','1 hotOutlier' , '5 coldOutlier']

# Create a dictionary that maps the original LISA labels to the new custom labels
labels = dict(zip(oldLabels, newLabels))

# Display the label-mapping dictionary
labels
```

```python
# Replace the original LISA cluster labels with the custom labels
# defined in the label-mapping dictionary.
lima.replace({'HS_lisa_queen':labels},inplace=True)

# Display the count of districts in each renamed LISA category,
# sorted by label order.
lima['HS_lisa_queen'].value_counts().sort_index()
```

```python
# Define a categorical color map with 5 classes for visualizing
# the Queen-based LISA cluster categories.
myColMap = plt.get_cmap('PuOr', 5)

# Plot the LISA cluster map using the renamed Queen LISA categories.
# Districts are colored by cluster type with boundaries outlined for clarity.
lima.plot(column='HS_lisa_queen',
                categorical=True,
                cmap=myColMap,
                linewidth=0.1,
                edgecolor='k',
                legend=True,
                legend_kwds={'bbox_to_anchor': (0.3, 0.3)},
                figsize=(12,12))
```

### **Spatial Autocorrelation**

**Moran's LISA with `knn8_lima`**

```python
# Compute Local Morans I (LISA) for the education variable using
# the K-nearest neighbors (k = 8) spatial weights.
lisa_knn8 = esda.Moran_Local(lima['Educ_sec_comp2019_pct'], knn8_lima)


# Plot the LISA cluster map using K-nearest neighbors (k = 8) weights,
# highlighting statistically significant clusters and outliers at the 0.05 level.
lisa_knn8.plot(lima, crit_value = 0.05, figsize = (10, 13), legend = True)
```

```python
# Get the LISA cluster (quadrant) labels for each district using
# the K-nearest neighbors (k = 8) weights at the 0.05 significance level.
lima['HS_lisa_knn8'] = lisa_knn8.get_cluster_labels(crit_value = 0.05)

# Display the count of districts in each LISA category
# for the KNN-based analysis.
lima['HS_lisa_knn8'].value_counts()
```

```python
# Define the summary statistics to compute for the education variable
TheStats = ["mean", "min", "max","var"]

# Calculate summary statistics of the education variable grouped by
# KNN-based (k = 8) LISA cluster category.
grouped_stats = lima.groupby('HS_lisa_knn8')['Educ_sec_comp2019_pct'].agg(TheStats)

# Display the summary statistics for each KNN-based LISA cluster group.
grouped_stats
```

```python
# Display the unique LISA cluster categories assigned under
# the K-nearest neighbors (k = 8) neighborhood definition.
lima['HS_lisa_knn8'].unique()
```

```python
# Define the original LISA cluster labels produced by the KNN-based analysis
oldLabels=['Insignificant', 'Low-Low', 'High-High', 'High-Low', 'Low-High']

# Define custom labels to rename the LISA categories for clarity and ordering
# (with 'no_pattern' placed in the middle).
newLabels = ['3 no_pattern', '4 coldSpot','2 hotSpot','1 hotOutlier' , '5 coldOutlier']

# Create a dictionary mapping the original labels to the new custom labels
labels = dict(zip(oldLabels, newLabels))

# Display the label-mapping dictionary
labels
```

```python
# Replace the original KNN-based LISA cluster labels with the custom labels
# defined in the label-mapping dictionary.
lima.replace({'HS_lisa_knn8':labels},inplace=True)

# Display the count of districts in each renamed KNN-based LISA category,
# sorted by label order.
lima['HS_lisa_knn8'].value_counts().sort_index()
```

```python
# Define a categorical color map with 5 classes for visualizing
# the KNN-based (k = 8) LISA cluster categories.
myColMap2 = plt.get_cmap('PuOr', 5)

# Plot the LISA cluster map using the renamed KNN-based LISA categories.
# Districts are colored by cluster type with boundaries outlined for clarity.
lima.plot(column='HS_lisa_knn8',
                categorical=True,
                cmap=myColMap2,
                linewidth=0.1,
                edgecolor='k',
                legend=True,
                legend_kwds={'bbox_to_anchor': (0.3, 0.3)},
                figsize=(12,12))
```
